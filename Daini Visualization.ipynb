```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read in combined data

master_df=pd.read_csv(r"combined.csv", error_bad_lines=False, low_memory=False)
video_list=master_df['video_id'].unique()

#Regenerate figures for plotting graphs

df=master_df[['video_id','title','views','video_likes','dislikes','comment_total','category']]
df.drop_duplicates(inplace=True)
df.reset_index(drop=True)
df.to_csv('figures.csv',index=False)

df=pd.read_csv(r"figures.csv", error_bad_lines=False, low_memory=False)
```

# 1. Plot category by number of views

```python
import matplotlib.pyplot as plt
import numpy as np

category_list=df['category'].tolist()
category_list=list(set(category_list))
for i in category_list:
    if str(i) == 'nan':
        category_list.remove(i)
print (category_list)

view_count=[]
for cat in category_list:
    tmp_df=df.loc[df['category']==cat]
    tmp_views=tmp_df['views'].tolist()
    tmp_sum=np.sum(tmp_views)
    view_count.append(tmp_sum)
    
print (view_count)

# Create a dataframe with sorted view numbers
d={'category':category_list,'views':view_count}
tmp_df=pd.DataFrame(data=d)
tmp_df.sort_values(by='views',ascending=False,inplace=True)

#Plot graph
category_list=tmp_df['category']
view_count=tmp_df['views']
plt.bar(category_list,view_count, align='center')
plt.xlabel('Video category',fontsize=18)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel('Views',fontsize=18)
plt.title('Category rank by views',fontsize=20)
fig =plt.gcf()
fig.set_size_inches(20, 10.5)
fig.savefig('by_views.png', dpi=200)
plt.show()
```

# 4. Plot video distribution by category in percentage

```python
category_list=df['category'].tolist()
category_list=list(set(category_list))
for i in category_list:
    if str(i) == 'nan':
        category_list.remove(i)
print (category_list)

totalnum=df.shape[0]

number_of_videos=[]
for cat in category_list:
    tmp_df=df.loc[df['category']==cat]
    tmp_count=tmp_df.shape[0]
    number_of_videos.append(tmp_count*100/totalnum)

print (number_of_videos)

d={'category':category_list,'video percentage':number_of_videos}
tmp_df=pd.DataFrame(data=d)

fig1, ax1 = plt.subplots()
ax1.pie(number_of_videos,labels=category_list, autopct='%1.2f%%',
        startangle=90,radius=1.5)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
fig =plt.gcf()
fig.set_size_inches(20, 10.5)
fig.savefig('video_percentage.png', dpi=200)
plt.show()
```
